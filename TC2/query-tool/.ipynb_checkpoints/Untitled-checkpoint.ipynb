```python
import numpy as np
from faiss import IndexFlatIP
from llama_cpp import Llama
from PyPDF2 import PdfReader
```

```python
reader = PdfReader("../data/pdf/UBER_Expert Call 1.pdf")
full_text = []
for page in reader.pages:
    full_text.append(page.extract_text() or "")
text = "\n".join(full_text)
print(text)
```

```python
transcript = text
max_tokens = 500
overlap = 100

words = transcript.split()
chunks = []
start = 0

while start < len(words):
    end = min(len(words), start+max_tokens)
    chunk = " ".join(words[start:end])
    chunks.append(chunk)
    if end == len(words): break
    start = end - overlap
```

```python
len(chunks), len(chunks[0])
```

```python
llm_emb = Llama(model_path = "./llama-2-7b.Q4_K_S.gguf", use_mmap=False, verbose=False, embedding=True, n_ctx=4096, n_threads=8)

def embed(st):
    raw_emb = llm_emb.create_embedding(st)
    emb = raw_emb["data"][0]["embedding"]
    return emb[0]
```

```python
embed("Hello")
```

```python
chunk_embs = np.vstack([embed(c) for c in chunks])
```

```python
dim    = chunk_embs.shape[1]
index  = IndexFlatIP(dim)           # flat index for maximum recall
index.add(chunk_embs)               # now holds all your chunk vectors
```

```python
import os

def get_files_in_directory(dir_path):
  try:
    files = [f for f in os.listdir(dir_path) if os.path.isfile(os.path.join(dir_path, f))]
    return files
  except FileNotFoundError:
    return f"Error: Directory not found: {dir_path}"
  except NotADirectoryError:
    return f"Error: Not a directory: {dir_path}"
  except Exception as e:
    return f"An error occurred: {e}"

# Example usage:
directory_path = "../data/text/"  # Replace with the actual path
file_list = get_files_in_directory(directory_path)
print(file_list)
```
